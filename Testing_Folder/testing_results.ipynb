```python
def check_training_df(df,year):
    
    sample = df.sample(50).copy()
    
    
    names_col = col.get_col_contains("Name",sample.columns)
    team_col = ["Tm","Opp"]
    
    filtered_sample = sample[names_col+team_col+["W/L"]]
    filtered_sample.reset_index(inplace=True)
    print(filtered_sample.head(2))
    
    inconcistencies = []
    
    for game in range(0,len(filtered_sample)):
        
        team = filtered_sample.iloc[game]["Tm"]
        if team == "None": continue
        print(get_key_from_dict( team_dict2 ,team))
        
        opponent = filtered_sample.iloc[game]["Opp"]
        if opponent == "None": continue
        print(get_key_from_dict( team_dict2 ,opponent))
        
        schedule_team = schedule_and_record(year, team)
        #print(schedule_team.head())
        baseball_date = datetime.strptime(filtered_sample.iloc[game]["Date"],"%Y-%m-%d").strftime("%A, %b %d").lstrip("0").replace(" 0", " ")
        team_game = schedule_team[schedule_team["Date"] == baseball_date]
        
        if team_game["Opp"].values == opponent: correct_opp = True
        else: correct_opp = False
            
        if team_game["W/L"].values == filtered_sample.iloc[game]["W/L"]: correct_result = True
        else: correct_result = False
            
        games_this_date = batting_stats_range(filtered_sample.iloc[game]["Date"],)
        #print(games_this_date)
        #print(set(games_this_date["Tm"]))
        
        players_team = games_this_date[games_this_date["Tm"]== get_key_from_dict( team_dict2 ,team) ]
        #print(players_team)
        
        players_opp = games_this_date[games_this_date["Tm"]== get_key_from_dict( team_dict2 ,opponent) ]
        #print(players_opp)
        
        team_inconsistencies = []
        
        for team_player in range(1,10):
            #print(players_team.Name)
            if filtered_sample.iloc[game][team_player] in list(players_team.Name) : continue
            else: team_inconsistencies.append(filtered_sample.iloc[game][team_player])
            #print(filtered_sample.iloc[game][team_player])
            #print(team_inconsistencies)
        
        opp_inconsistencies = []
        
        for opp_player in range(10,19):
            #print(players_opp.Name)
            if filtered_sample.iloc[game][opp_player] in list(players_opp.Name) : continue
            else: opp_inconsistencies.append(filtered_sample.iloc[game][opp_player])
            #print(filtered_sample.iloc[game][opp_player])
            #print(opp_inconsistencies)
        
        inconcistencies.append( { "Date" : filtered_sample.iloc[game]["Date"],
                                 "Team": get_key_from_dict( team_dict2 ,team),
                                 "From_Team" : team_inconsistencies,
                                 "Opponent":get_key_from_dict( team_dict2 ,opponent),
                                 "From_Opp" : opp_inconsistencies,
                                 "Correct_Opponent?": correct_opp,
                                 "Correct_Result?": correct_result
                                })

    return inconcistencies
```
